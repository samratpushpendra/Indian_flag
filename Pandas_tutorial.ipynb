## Groupby
Three stages


*   Split - we split dataframe into multiple smaller dataframe based on the values of keys
*   Apply - we apply desired aggregation/transformation on each dataframe
* Combine - we combine results from apply state into a datarame






```python
df = pd.DataFrame({'key': ['A','B','C']*2,
                   'data': range(6)})
```

```python
df
```

```python
df.groupby("key").sum()
```
